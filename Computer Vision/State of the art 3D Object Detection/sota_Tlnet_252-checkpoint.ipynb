```python
filename = "RCNN/000278.txt"
def project_ground_truth(img_gt):
    with open(filename, 'r') as f:
            fin = f.readlines()
         #   objectclasslist = []
            for line in fin:
                if line.split(" ")[0] != "DontCare":
                    #print(line)
                    x1_value = int(float(line.split(" ")[4]))
                    y1_value = int(float(line.split(" ")[5]))
                    x2_value = int(float(line.split(" ")[6]))
                    y2_value = int(float(line.split(" ")[7]))
                    distance = float(line.split(" ")[13])
                    object_class = str(line.split(" ")[0])
                   # objectclasslist.append(objectclasslist)
                    cv2.rectangle(img_gt, (x1_value, y1_value), (x2_value, y2_value), (0,205,0), 2)
                    cv2.putText(img_gt, str(distance), (int((x1_value+x2_value)/2),int((y1_value+y2_value)/2)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2, cv2.LINE_AA)
                    cv2.putText(img_gt, str(object_class), (int(x1_value),int(y1_value)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2, cv2.LINE_AA)
    return img_gt
imgL = cv2.imread('ssd/input/278/im0.png')
imgR = cv2.imread('ssd/input/278/im1.png')

img_gt = imgL.copy()
img_label = project_ground_truth(img_gt)
plt.figure(figsize=(20,40))
plt.imshow(img_gt)
plt.show()
```

```python
filename = "RCNN/000278.txt"
def get_depth(filename):
    gt_depthcl = []
#def read_depth_and_class_from_sota(STATEOFTHEART):
    with open(filename, 'r') as f:
            fin = f.readlines()
            for line in fin:
                if line.split(" ")[0] != "DontCare":
                    #access and copy item from (line)
                    distance = float(line.split(" ")[13])
                    classofobject = str(line.split(" ")[0])
                    collect = [classofobject, distance]
                    gt_depthcl.append(collect)
    return gt_depthcl
rcnn_depthcl = get_depth(filename)            
print(rcnn_depthcl)     
rcnn_depthcl = np.array(rcnn_depthcl)      #convert to array
rcnn_depthcl = rcnn_depthcl.reshape(len(rcnn_depthcl),2) #then reshape n by 2
print(rcnn_depthcl)
#gt 1342
#here 1324
###reorder 
#rcnn_depthcl = rcnn_depthcl[[0,1,3,2],:]
#print(rcnn_depthcl)
```

```python
#Repeat for groundtruth
filenamegt = "label_2/000278.txt"
def get_depth(filename):
    gt_depthcl = []
#def read_depth_and_class_from_groundtruth(ground_truth_text_file):
    with open(filename, 'r') as f:
            fin = f.readlines()
            for line in fin:
                if line.split(" ")[0] != "DontCare":
                    #access and copy item from (line)
                    distance = float(line.split(" ")[13])
                    classofobject = str(line.split(" ")[0])
                    collect = [classofobject, distance]
                    gt_depthcl.append(collect)
    return gt_depthcl
gt_depthcl = get_depth(filenamegt)            
print(gt_depthcl)     
gt_depthcl = np.array(gt_depthcl)      #convert to array
gt_depthcl = gt_depthcl.reshape(len(gt_depthcl),2) #then reshape n by 2
print(gt_depthcl)
```

```python
def collectresults(groundtruth, detection):
    dzframe = np.concatenate((groundtruth, detection), axis = 1)
    print(dzframe)
    print(dzframe.shape)
    df = pd.DataFrame(data=dzframe, columns=["classes", "real depth in m", "class", "detected depth in m"])
    df.head(11)
    df = df.drop(columns=["class"],axis=1)
    df.info()
    #ff=df.apply(pd.to_numeric)
    #ff.info
    df.head(11)
    df = df.sort_values(["classes","real depth in m"], ascending = True)
    df.head(11)
    R = df["real depth in m"]
    D = df["detected depth in m"]

    #R.astype("float") #alternative or df["xxx"].astype("e.g int")
    #converting the datatype to float (easier than whole dataset)
    #try category method to reduce mem occupied space R[xx].astype("category")
    R= R.apply(pd.to_numeric)
    D= D.apply(pd.to_numeric)
    R.dtype
    D.dtype
    Num = R.sub(D)
    dz = Num.div(R)    #(real - detected)/real

    dz.head(11)

    abs_dz =dz.abs()    #(real - detected)/real



    dz.head(11)

    print(abs_dz)

    #reassigning values to df

    df["real depth in m"]= R
    df["detected depth in m"]= D

    #inserting new colums #method .insert

    df["relative error"] = dz
    df["abs relative error"] = abs_dz

    df.head(11)

    return df, R, D

df, R, D = collectresults(gt_depthcl,rcnn_depthcl)
df.head(11)

```
